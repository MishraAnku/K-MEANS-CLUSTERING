## Traditional k-means utilizes a randomization process for initializing these centroids -- though this is not the only approach -- but poor initialization can lead to increased numbers of required clustering iterations to reach convergence, and a less-efficient algorithm overall. k-means++ is relatively consistent within clusters, with the specific data points selected as initial centroids varying slightly.<a href="#Traditional-k-means-utilizes-a-randomization-process-for-initializing-these-centroids----though-this-is-not-the-only-approach----but-poor-initialization-can-lead-to-increased-numbers-of-required-clustering-iterations-to-reach-convergence,--and-a-less-efficient-algorithm-overall.-k-means++-is-relatively-consistent-within-clusters,-with-the-specific-data-points-selected-as-initial-centroids-varying-slightly." class="anchor-link">¶</a>

    # We create 10 clusters initailly
    # here we are importing the kmeans module from the sklearn.cluster pacakge 
    # ideally the value of k means intial point should be odd.

    from sklearn.cluster import KMeans

    # wcss is initialised as an empty list , which will be populated with the wcss of
    # each cluster as the loop iterates
    # here we have  created the global valiable which is wcss  , which is the empty list 
    # once the wcss will be populated with the value  then it  the list will be filled up with values 
    # then the list fills up .

    wcss = []   

    # for 10 clusters,the upper bound is 11,as the range will include 1 less than the upper bound limit
    # here ideally we will get the 10 clusters 

    for i in range(1, 11):   

    # hre k means is the object 
    # the object is created from the class and the class name is k means 
    # we works with the object not with the class 
    # here we are not applying the traditional k means clustering but here we are applying the
    # k means ++ clustering  which will be giving the optimal numbers of the clusters .
    # random state ensures the permanently .
    # random state value depends on the choice 
    # if we will execute the code 100 times , 100 times the resulrt will be same 
    # n_clusters = i - it means number of clusters = i that means i = 10  that means the initial 
    # number of clusters is 10 
    # if the optimal number of clusters and the initial number of clusters  are close to each other
    # then it is fine or if  not go with the other numbers of clusters 
    # the value should be very close to the optimal number of clusters .

        kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
        
    # Fit method helps to run the loop/algorithmn on the training set
    # here we are fitting the x feature to the  k means

        kmeans.fit(X) 
        
    # inertia is the parameter of the object kmeans which helps us generate wcss
    # here we have created the object called wcss  and we have created it as the empty list 
    # we have appended the kmeans as an inertia .
    # this inertia will create the wcss 
    # then the wcss is generated 

        wcss.append(kmeans.inertia_) 

    # we start this line shifting indent to left,to confirm that we are out of loop
    # here we are plotting the plot  having the x and y axis 
    # we are having the range of 1 to 11 on the x axis and  having the wcss on the y axis .

        plt.plot(range(1, 11), wcss) 

    # range 1-11 shows what reflects in X axis,while wcss reflects in y axis
    # here we are plotting the title of the graph 

        plt.title('The Elbow Method')  
        
    # then we are giving the label of the x label and the y label 
        
        plt.xlabel('Number of clusters')

        plt.ylabel('WCSS')
        
    # then we are plotting to show the graph 

        plt.show()
